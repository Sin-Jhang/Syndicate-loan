```python
import os
import numpy as np
import pandas as pd

import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.tokenize import regexp_tokenize
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
article_11_15 = os.listdir('/Users/sinjhanghuang/Documents/碩士畢業論文/柏凱聯合論文/柏凱合作文章2011-2015（整合版）/2011-2015')
len(article_11_15)
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
all_data = pd.read_excel('/Users/sinjhanghuang/Documents/碩士畢業論文/柏凱聯合論文/柏凱合作文章2011-2015（整合版）/2011-2015 company_name_clear.xlsx')
all_data
```

```python
for i in range(0, len(article_11_15)):
    data = pd.read_csv('/Users/sinjhanghuang/Documents/碩士畢業論文/柏凱聯合論文/柏凱合作文章2011-2015（整合版）/2011-2015/{}'\
                       .format(article_11_15[i]))
    data = data[(data['head_related'] > 0) | (data['text_related'] > 0)]
    data = data.reset_index(drop=True)
    
    print(i,'事件: ',article_11_15[i][:6],' 正在處理中...')
    
    # head
    data['sentiment_score_head'] = data['head'].apply(lambda x: sia.polarity_scores(x)["compound"])
#     data['sentiment_head'] = np.select([data["sentiment_score_head"] < 0, data["sentiment_score_head"] == 0, data["sentiment_score_head"] > 0],
#                                   [-1, 0, 1])
    
    # text
    data['sentiment_score_text'] = data['text'].apply(lambda x: sia.polarity_scores(x)["compound"])
#     data['sentiment_text'] = np.select([data["sentiment_score_text"] < 0, data["sentiment_score_text"] == 0, data["sentiment_score_text"] > 0],
#                                   [-1, 0, 1])
    data['company'] = all_data['company_clear'][all_data['FacilityID']==int(article_11_15[i][:6])].values[0]
    
    data = data[['company','date','head','text','head_related','text_related','sentiment_score_head','sentiment_score_text']]
    
    
    if len(data) != 0:
        data.to_csv('/Users/sinjhanghuang/Documents/碩士畢業論文/柏凱聯合論文/柏凱合作文章2011-2015（整合版）/2011-2015 sentiment/{}'\
                    .format(article_11_15[i]), index=False)
    else:
        continue
    print(i,'事件: ',article_11_15[i][0:6],' 已完成')
```

## 計算所有senitment的平均數

```python
article = os.listdir('/Users/sinjhanghuang/Documents/碩士畢業論文/柏凱聯合論文/柏凱合作文章2011-2015（整合版）/2011-2015 sentiment')
len(article)
```

```python
score = 0
number = 0
for i in range(0, len(article)):
    data = pd.read_csv('/Users/sinjhanghuang/Documents/碩士畢業論文/柏凱聯合論文/柏凱合作文章2011-2015（整合版）/2011-2015 sentiment/{}'\
                       .format(article[i]))
    for x in range(0, data.shape[0]):
        sentiment = data.loc[x, 'sentiment_score_text']
        score += sentiment
        number += 1
```
